```python
from numpy import loadtxt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU,PReLU,ELU
from keras.layers import Dropout
import pandas as pd
import matplotlib.pyplot as plt
import csv
import numpy as np
import seaborn as sns
```

```python
data_path = '/content/drive/MyDrive/ml_datasets/nn_data.txt'
dataset = pd.read_csv(data_path, header=None)
dataset.shape
print(dataset.head())
```

```python
X = dataset.iloc[:,0:8]
y = dataset.iloc[:,8]

X.head()
```

```python
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)


X_normalized
```

```python
model = Sequential()
model.add(Dense(8, input_shape=(8,), activation='relu'))
model.add(Dense(6, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X, y, epochs=100, batch_size=10)
```

```python
_, accuracy = model.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))
```

```python
# fit the keras model on the dataset for normalized data
model.fit(X_normalized, y, epochs=100, batch_size=10)
```

```python
_, accuracy = model.evaluate(X_normalized, y)
print('Accuracy: %.2f' % (accuracy*100))
```

```python
# make class predictions with the model
predictions = (model.predict(X) > 0.5).astype(int)
# summarize the first 5 cases
for i in range(5):
	print(' %d (expected %d)' % ( predictions[i], y[i]))
```
